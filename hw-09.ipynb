```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

Data_2d

```python
data_2d = pd.read_csv('data_2d.csv')
X_2d = data_2d.iloc[:, 1:3] 
```

```python
def elbow_method(X, k_range):
    ssd = []
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=0)
        kmeans.fit(X)
        ssd.append(kmeans.inertia_)
    plt.figure(figsize=(10, 6))
    plt.plot(k_range, ssd, marker='o')
    plt.title('Elbow Method For Optimal k')
    plt.xlabel('Number of clusters')
    plt.ylabel('Sum of squared distances')
    plt.grid(True)
    plt.show()

elbow_method(X_2d, range(1, 11))
```

Desided to make 5 clusters

```python
def perform_kmeans(X, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(X)
    plt.figure(figsize=(10, 6))
    plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', edgecolor='black')
    centers = kmeans.cluster_centers_
    plt.scatter(centers[:, 0], centers[:, 1], s=250, marker='*', c='red', edgecolor='black', label='Centroids')
    plt.title(f'K-Means Clustering with {n_clusters} Clusters')
    plt.xlabel('Dimension 1')
    plt.ylabel('Dimension 2')
    plt.legend()
    plt.grid(True)
    plt.show()


perform_kmeans(X_2d, 5)
```

mnist.csv

```python
from sklearn.decomposition import PCA
```

```python
mnist = pd.read_csv('mnist.csv')
mnist_data = mnist.iloc[:, 1:]
```

```python
pca = PCA(n_components=0.95) 
mnist_reduced = pca.fit_transform(mnist_data)
```

```python
def elbow_method(X, k_range):
    ssd = []
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=0)
        kmeans.fit(X)
        ssd.append(kmeans.inertia_)
    plt.figure(figsize=(10, 6))
    plt.plot(k_range, ssd, marker='o')
    plt.title('Elbow Method For Optimal k (PCA Reduced MNIST)')
    plt.xlabel('Number of clusters')
    plt.ylabel('Sum of squared distances')
    plt.grid(True)
    plt.show()

elbow_method(mnist_reduced, range(1, 11))
```

decided to make 9 clusters

```python
def perform_kmeans(X, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(X)
    plt.figure(figsize=(10, 6))
    plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', edgecolor='black')
    centers = kmeans.cluster_centers_
    plt.scatter(centers[:, 0], centers[:, 1], s=250, marker='*', c='red', edgecolor='black', label='Centroids')
    plt.title(f'K-Means Clustering on PCA-Reduced MNIST with {n_clusters} Clusters')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend()
    plt.grid(True)
    plt.show()


perform_kmeans(mnist_reduced, 9) 
```
